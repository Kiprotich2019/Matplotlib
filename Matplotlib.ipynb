```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x**2
y
```

```python
plt.plot(x,y,'ro',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Diagram')
```

```python
num_subplots = 4

# Create a 2x2 subplot layout
plt.figure(figsize=(8, 6))

# Loop through each subplot
for i in range(1, num_subplots + 1):
    plt.subplot(2, 2, i)
    #plt.plot(x, y, f'{i}--')
    plt.plot(x, y,linestyle='dashed',label=f'Line {i}')
    plt.xlabel(f'X-axis Label {i}')
    plt.ylabel(f'Y-axis Label {i}')
    plt.legend()

# Adjust layout for better spacing
plt.tight_layout()
```

```python
##creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.xlabel('X-axis Label 1')
plt.ylabel('Y-axis Label 1')

plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.xlabel('X-axis Label 2')
plt.ylabel('Y-axis Label 2')

plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.xlabel('X-axis Label 3')
plt.ylabel('Y-axis Label 3')

plt.subplot(2,2,4)
plt.plot(x,y,'y--')
plt.xlabel('X-axis Label 4')
plt.ylabel('Y-axis Label 4')

plt.tight_layout()
```

```python
#pie chart
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice
#Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)
plt.axis('equal')
plt.show()
```
